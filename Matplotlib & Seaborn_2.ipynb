# CDF & PDF 

Difference between Cummulative distribution function & Probability distribution function.

```python
counts, bin_edges = np.histogram(df['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))

print(pdf)
print(bin_edges)

cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.xlabel('Petal Length')
```

Blue Line - Probability Distribution Function <br>
Red Line - Cummulative Distribution Function <br>

Cummulative Frequency Distribution <br>
Suppose take an example of 4 (Petal Length) - Then according to CDF it means there are atleast 41% (Y axis) of flowere having Petal length less than equal to 4. It is a cummulative of every related elements.

Probability Density Function <br>
It says like for example if we take 4.5 & 5 points on x axis, it mean approx 19% flowers have petal length between 4.5 to 5. It shows total count of flowers in each categoty.
