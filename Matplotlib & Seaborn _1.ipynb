# Matplotlib Tutorial

Matplotlib is a plotting library for the python programming language and its numerical mathematics extension Numpy. It provides an object oriented API for embedding plots into applications using general-puropose GUI toolkits like Tkinter, wxPython, QT or GTK+.

Some of the major pros of Matplotlibs are: <br>
Generally easy to get started for simple plots <br>
Support for custom labels and texts <br>
Great control of every element in a figure <br>
High-quality output in many formats <br>
very customizable in general <br>

```python
import matplotlib.pyplot as plt
%matplotlib inline  
#Due to this line we do not need to use pyplot.show()
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

# Plotting using matplotlib


## Scatter Chart

```python
plt.scatter(x,y,c='green')
plt.xlabel('x-axis')
plt.ylabel('-axis')
plt.title('Relation b/w X & Y')
plt.savefig('Test.png')
plt.show()
```

## Plot Chart - Line Chart

```python
y=x*x
plt.plot(x,y,color='red',linestyle='--',linewidth=2,marker='^',markersize=10)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2d Diagram')
```

## Sub-Plots 

```python
#plt.subplot(no. of rows,no. of columns, index)

plt.subplot(2,2,1)
plt.plot(x,y,color='g',linestyle='-',linewidth=2,marker='^',markersize=10)
plt.subplot(2,2,2)
plt.plot(x,y,color='r',linestyle='--',linewidth=2,marker='o',markersize=10)
plt.subplot(2,2,3)
plt.plot(x,y,color='b',linestyle='-.',linewidth=2,marker='v',markersize=10)
plt.subplot(2,2,4)
plt.plot(x,y,color='y',linestyle=':',linewidth=2,marker='1',markersize=10)
```

```python
x=np.arange(1,11)
y=(x*30)

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
np.pi
```

```python
#Compute the x & y coordinates for points on a single curve
x = np.arange(0,4*np.pi,0.1)
y= np.sin(x)

plt.plot(x,y)
plt.title('Sine wave form')
plt.show()
```

## Bar Chart

```python
x=[2,8,10]
y=[11,16,19]

x2=[3,9,11]
y2=[6,15,17]

x3 = [2,8,9]
y3=[5,6,9]

plt.bar(x,y,color='y')
plt.bar(x2,y2,color='g')
plt.bar(x3,y3,color='r')
plt.title('Bar Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

## Histogram

```python
a=np.array([22,45,6,7,8,9,0,])
plt.hist(a,bins=15)
plt.title('Histogram')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

# Boxplot 

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()
```

# Pie Charts

```python
labels=['Python','C++','Ruby','Julya']
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.4,0,0,0)

plt.pie(sizes,explode=explode,colors=colors,labels=labels,autopct='%1.1f%%',shadow=True)
plt.show()
```

## Univariate & Bivariate Analysis

If we have 2 features then it is called univariate analysis. If we have 3 features then it is called bivariate analysis. <br>
distplot is mainly use when we have one feature in our dataset.  <br>
joinplot is mainly use when we have two features in our dataset. <br>
pairplot is mainly use when we have more than two features in our dataset.

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

# Correlation with Heatmap 

A correlation heatmap uses colored cells, typically in a monochronic scale, to show a 2D correlation matrix(table) between two discrete dimensions or event types. It is very important in feature selection.

Correlation can only find out when there are int or float values.

```python
df.corr()
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())
```

# Joinplot 

A joint plot allows to study the relationship between 2 numeric varaibles. The central chart display their correlation.It is usually a scatterplot, a hexibin plot, a 2D histogram or a 2D density plot

## Univariate Analysis 

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
#Probability Density function
```

# Pairplot 

A pairplot is also known as scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

# Distplot 

Distplot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'],kde=True)

```

```python
df['tip'].value_counts()
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
#Kernel density estimulation - KDE
```
